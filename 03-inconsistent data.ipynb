```python
import pandas as pd
data = {
    'date': ['2024-12-01', '01-12-2023', '2024/12/01', '12-01-2023'],
    'country': ['USA', 'U.S.A.', 'America', 'United States'],
    'name': ['Kainat', 'Kianat', 'Amara', 'Ammar'],
    'sales_2023': [100, 200, None, 200],
    'sales_2024': [None, 150, 300, 150]
}
# make pandas dataframe
df = pd.DataFrame(data)
df.head()
```

```python
# standardizing the date format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# Harmonize the name of the coutry
country_mapping = {'USA': 'United States', 'U.S.A.': 'United States', 'America': 'United States'}
df['country'] = df['country'].replace(country_mapping)
df.head()
```

```python
# Correct the typographical Mistakes in name
df['name'] = df['name'].replace({'Kianat': 'Kainat', 'Amara': 'Ammar'})
df.head()
```

```python
# remove duplicates
df = df.drop_duplicates(subset="name")
df.head()
```
